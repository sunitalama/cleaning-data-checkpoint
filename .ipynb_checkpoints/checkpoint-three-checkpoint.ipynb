```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import plotly.express as px


# read the data

df = pd.read_csv(r"C:\Users\lama_\Downloads\metadata.csv") 
df.head()
```

## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
# check for missing values
df.isnull().sum()
```

```python
# % of missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# overall percentage of missing data from the dataset
total_cells = np.product(df.shape)

missing_cells = pd.isnull(df).sum()

total_missing = missing_cells.sum()

percentage_missing = round((total_missing/total_cells), 2) *100

print(percentage_missing, "% Missing cells from the data")    
```

```python
# heatmap to see missing data
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
#filling na values for categorial data
df["Year"] = df["Year"].fillna("unknown")
```

```python
df["Country"] = df["Country"].fillna("x")
```

```python
df["Continent"] = df["Continent"].fillna("Y")
```

```python
# # Updating the values in 'azm_sr', 'cip_sr', 'cro_sr', 'cfx_sr', where 0 = "Sensitive" and 1 = "Resistance"
df = df.replace({'azm_sr': {0: "Sensitive", 1: "Resistant"},
                'cip_sr': {0: "Sensitive", 1: "Resistant"},
                'cro_sr': {0: "Sensitive", 1: "Resistant"},
                'cfx_sr': {0: "Sensitive", 1: "Resistant"}}
                 )
```

```python
# Renaming columns for easier comprehension, #mic=minimum inhibitory concentration
df = df.rename(columns={"azm_mic": "Azithromycin mic", 
                        "cip_mic": "Ciprofloxacin mic", 
                        "cro_mic": "Ceftriaxone mic", 
                        "cfx_mic": "Cefixime mic", 
                        'log2_azm_mic':'Azithromycin log mic',
                        'log2_cip_mic':'Ciprofloxacin log mic',
                        'log2_cro_mic':'Ceftriaxone log mic',
                        'log2_cfx_mic':'Cefixime log mic',
                        'azm_sr': 'Azithromycin sr',
                        'cip_sr': 'Ciprofloxacin sr',
                        'cro_sr': 'Ceftriaxone sr',
                        'cfx_sr': 'Cefixime sr'})
```

## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

```python
# Redundant mic for all antibiotics are dropped.
#columns with more than 61% of missing values are dropped
# Remaining columns are not necessary for our study.
to_drop = ['Azithromycin', 'Ciprofloxacin', 'Ceftriaxone', 'Cefixime','Tetracycline', 'Penicillin', 
           'Group','Beta.lactamase','NG_MAST', 'Group',  'tet_mic', 'pen_mic',  'log2_tet_mic',
          'log2_pen_mic', 'tet_sr',
          'pen_sr']

df.drop(columns= to_drop, inplace=True)
```

```python
df.head()
```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
# These columms  should contain either 'Resistance' or 'Sensitive'.  dropping all NaN values is the best idea.

df.dropna(subset=['Azithromycin sr',
                 'Ciprofloxacin sr',
                'Ceftriaxone sr',
                'Cefixime sr'], inplace= True)


```

```python
df.isnull().sum()
```

```python
# After dropping Nan values from above 4 columns, there is  no remaining misssing values, so as predicted before we dont have to fill the missing values of other numerical columns with aggregated mean.
```
